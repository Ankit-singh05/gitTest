```python
raw_path = "/Volumes/databricks_training/delta_demo/raw_data/customers/"

raw_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(raw_path)
)

raw_df.display()
raw_df.printSchema()
```

```python
r_path = "/Volumes/databricks_training/delta_demo/raw_data/orders/"

r_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .parquet(r_path)
)

r_df.display()
r_df.printSchema
```

```python
ro_path = "/Volumes/databricks_training/delta_demo/raw_data/products/"

ro_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .parquet(ro_path)
)

ro_df.display()
ro_df.printSchema()
```

```python
bronze_path = "/Volumes/databricks_training/delta_demo/bronze/customer_bronze"

(
    raw_df.write
    .format("delta")
    .mode("overwrite")
    .save(bronze_path)
)
```

```python
bronze_path = "/Volumes/databricks_training/delta_demo/bronze/order_bronze"

(
    r_df.write
    .format("delta")
    .mode("overwrite")
    .save(bronze_path)
)
```

```python
bronze_path = "/Volumes/databricks_training/delta_demo/bronze/product_bronze"

(
    ro_df.write
    .format("delta")
    .mode("overwrite")
    .save(bronze_path)
)
```
