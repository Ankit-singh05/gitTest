```python
silver_path = "/Volumes/databricks_training/delta_demo/silver/customers_silver/"
silver_df = spark.read.format("delta").load(silver_path)
```

```python
from pyspark.sql.functions import col, to_date, regexp_replace ,coalesce, lit,trim,upper , count

customers_gold_df = (
    silver_df
    .groupBy("geography_id")
    .agg(count("customer_id").alias("total_customers"))
)

```

```python
customers_gold_path = "/Volumes/databricks_training/delta_demo/gold/customers_by_geography"

(
    customers_gold_df.write
    .format("delta")
    .mode("overwrite")
    .save(customers_gold_path)
)
```

```python


products_silver_df = spark.read.format("delta").load(
    "/Volumes/databricks_training/delta_demo/silver/product_silver/"
)
```

```python
from pyspark.sql.functions import sum

products_gold_df = (
    products_silver_df
    .withColumn("profit", col("price") - col("cost"))
    .groupBy("category")
    .agg(
        sum("price").alias("total_revenue"),
        sum("cost").alias("total_cost"),
        sum("profit").alias("total_profit")
    )
)
```

```python
products_gold_path = "/Volumes/databricks_training/delta_demo/gold/product_profit_summary"

(
    products_gold_df.write
    .format("delta")
    .mode("overwrite")
    .save(products_gold_path)
)
```
