```python
raw_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/Volumes/demo_catalog/delta_test/filestore/orders_dataset.csv")

raw_df.display(5)
raw_df.printSchema()
```

```python
base_path = "/Volumes/demo_catalog/delta_test/filestore/delta"

dbutils.fs.mkdirs(f"{base_path}/bronze")
dbutils.fs.mkdirs(f"{base_path}/silver")
dbutils.fs.mkdirs(f"{base_path}/gold")
```

```python
raw_df.write \
    .format("delta") \
    .mode("overwrite") \
    .save(f"{base_path}/bronze/orders")

```

```python
bronze_df = spark.read.format("delta") \
    .load(f"{base_path}/bronze/orders")

bronze_df.display(5)
bronze_df.printSchema()
```

```python
from pyspark.sql.functions import col, to_timestamp

silver_df = (
    bronze_df
    .dropDuplicates(["order_id"])                     # remove duplicates
    .withColumn("order_date", to_timestamp(col("order_date")))  
    .withColumn("amount", col("amount").cast("double"))
    .filter(col("amount").isNotNull())
)

silver_df.show(5)
silver_df.printSchema()
```
